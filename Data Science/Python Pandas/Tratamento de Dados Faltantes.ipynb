## Continuação do tratamento dos dados faltantes

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados.fillna(0, inplace = True)
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados.info()
```
